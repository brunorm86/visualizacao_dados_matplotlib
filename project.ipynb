# Aula 1 - Conhecendo a biblioteca Matplotlib

```python
import pandas as pd

df = pd.read_csv('data/imigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
type(brasil)
```

```python
brasil_dic = {'ano':brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
brasil_dic
```

```python
dados_brasil = pd.DataFrame(brasil_dic)
```

```python
dados_brasil
```

```python
type(dados_brasil)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(dados_brasil['ano'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.xticks(rotation=45)
```

```python
plt.figure(figsize = (10,6))
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel("Ano")
plt.ylabel("Imigrantes")
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show() #para nao aparecer o texto acima do gráfico
```
